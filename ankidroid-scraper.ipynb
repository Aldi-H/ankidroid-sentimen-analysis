```python
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
```

```python
APP_ID = "com.ichi2.anki"
LANG = "en"
COUNTRY = "us"
TARGET_COUNT = 149759
BATCH_SIZE = 200
OUTPUT_FILE = "ankidroid_reviews.csv"
```

```python
def scraper_reviews(app_id, lang, country, target, batch_size):
    all_reviews = []
    token = None
    batch_num = 0

    while len(all_reviews) < target:
        batch_num += 1

        try:
            if token is not None:
                batch, token = reviews(
                    app_id,
                    lang=lang,
                    country=country,
                    sort=Sort.MOST_RELEVANT,
                    count=batch_size,
                    continuation_token=token 
                )
            else:
                batch, token = reviews(
                    app_id, lang=lang,
                    country=country,
                    sort=Sort.MOST_RELEVANT,
                    count=batch_size
                )

        except Exception as ex:
            print(f"    Error on batch {batch_num}: {ex}")
            continue

        if not batch:
            print("No more reviews")
            break

        all_reviews.extend(batch)

        if token is None:
            print("last page")
            break

    return all_reviews
```

```python
def save_to_csv(records, filepath):
    df = pd.DataFrame(records)

    df.to_csv(filepath, index=False, encoding="utf-8-sig")

    return df
```

```python
if __name__ == "__main__":
    ankidroid_scraper_reviews = scraper_reviews(
        app_id=APP_ID,
        lang=LANG,
        country=COUNTRY,
        target=TARGET_COUNT,
        batch_size=BATCH_SIZE
    )

    df = save_to_csv(ankidroid_scraper_reviews, OUTPUT_FILE)
```
